# Eventos de Mouse y Teclado en JavaScript

## 1. `mouseover()`
El evento `mouseover` se activa cuando el puntero del mouse entra en el área de un elemento. Este evento se propaga a través de los elementos padre, lo que significa que también se activa en los elementos superiores.

### Ejemplo:
        
        element.addEventListener('mouseover', function() {
            console.log('El mouse está sobre el elemento.');
        });
        
2. mouseout()

- El evento mouseout se activa cuando el puntero del mouse sale del área de un elemento.
- Al igual que mouseover, este evento se propaga hacia los elementos padre.

- Ejemplo:

        element.addEventListener('mouseout', function() {
            console.log('El mouse ha salido del elemento.');
        });
  
3. hover()

- El método hover no es un evento nativo de JavaScript, sino que es común en jQuery.
- Se utiliza para aplicar estilos cuando el mouse está sobre un elemento y revertir esos estilos cuando el mouse sale del mismo.

- Ejemplo en jQuery:
        
        
        $(element).hover(
            function() {
                $(this).css('background-color', 'yellow');
            }, 
            function() {
                $(this).css('background-color', 'white');
            }
        );
4. click

- El evento click se activa cuando un elemento es presionado y liberado con el botón del mouse.

- Ejemplo:

        element.addEventListener('click', function() {
            console.log('Elemento clicado.');
        });
  
5. dblclick
- El evento dblclick se activa cuando un elemento es clicado dos veces rápidamente.

- Ejemplo:

        element.addEventListener('dblclick', function() {
            console.log('Elemento doble clicado.');
        });
  
6. focus

- El evento focus se activa cuando un elemento (como un campo de entrada) recibe el foco, es decir, cuando el usuario hace clic en él o lo selecciona.

- Ejemplo:

        inputElement.addEventListener('focus', function() {
            console.log('El campo ha recibido el foco.');
        });
  
7. blur

- El evento blur se activa cuando un elemento pierde el foco, lo que significa que el usuario ha hecho clic fuera de él o ha seleccionado otro elemento.

- Ejemplo:

        inputElement.addEventListener('blur', function() {
            console.log('El campo ha perdido el foco.');
        });
  
8. mousedown()

- El evento mousedown se activa cuando el botón del mouse se presiona sobre un elemento.

- Ejemplo:

        element.addEventListener('mousedown', function() {
            console.log('Botón del mouse presionado.');
        });

  
9. mouseup()

- El evento mouseup se activa cuando el botón del mouse se libera después de haber estado presionado.

- Ejemplo:
        
        element.addEventListener('mouseup', function() {
            console.log('Botón del mouse liberado.');
        });
  
10. mousemove()

- El evento mousemove se activa cuando el puntero del mouse se mueve sobre un elemento.

- Ejemplo:

        element.addEventListener('mousemove', function(event) {
            console.log(`Coordenadas: X: ${event.clientX}, Y: ${event.clientY}`);
        });


- Este formato proporciona una visión clara y organizada de los eventos mencionados en JavaScript.
- Puedes usarlo para referencia o en tu documentación.
