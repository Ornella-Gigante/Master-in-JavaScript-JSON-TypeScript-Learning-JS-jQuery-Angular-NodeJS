# Plugins y widgets en JavaScript

## 1. Tooltip

- ¿Qué es?

- Un tooltip es un cuadro pequeño que aparece cuando pasas el cursor sobre un elemento, como un botón o enlace.

- ¿Qué hace?

- Muestra un mensaje breve o información adicional sin hacer clic en el elemento.

- ¿Para qué sirve?

- Es útil para dar detalles o instrucciones rápidas sin ocupar espacio en la pantalla.

#### Ejemplo de uso:


        $("#miElemento").tooltip();

        
## 2. Cuadros de diálogo (Popup)

- ¿Qué es?

- Un cuadro de diálogo o popup es una ventana pequeña que aparece encima del contenido de la página.

- ¿Qué hace?
- Interrumpe la interfaz para mostrar un mensaje importante o para pedirle al usuario que realice una acción (como confirmar o cancelar).

- ¿Para qué sirve?
- Es útil para alertas, confirmaciones, formularios rápidos, o cualquier información que requiera la atención inmediata del usuario.

### Ejemplo de uso:


        $("#miDialogo").dialog();

        
## 3. dialog()

- ¿Qué es?
- dialog() es un método de jQuery UI que crea un cuadro de diálogo interactivo.

- ¿Qué hace?

- Convierte un elemento (como un <div>) en una ventana de diálogo que el usuario puede mover, cerrar, y que puede tener botones como "Aceptar" o "Cancelar".

- ¿Para qué sirve?
- Se usa para mostrar formularios, alertas o mensajes de confirmación dentro de una ventana que el usuario puede manipular fácilmente.

#### Ejemplo de uso:

        
        $("#miElemento").dialog({
          title: "Título del diálogo",
          buttons: {
            Aceptar: function() {
              $(this).dialog("close");
            }
          }
        });

## 4. datepicker()

- ¿Qué es?
- datepicker() es un calendario interactivo que permite al usuario seleccionar una fecha.

- ¿Qué hace?
- Muestra un pequeño calendario en el cual el usuario puede hacer clic para elegir una fecha. La fecha seleccionada se coloca en un campo de texto.

- ¿Para qué sirve?
- Es muy útil para formularios donde se necesita que el usuario seleccione una fecha, como en reservas, agendas o citas.

#### Ejemplo de uso:


        $("#miCampoFecha").datepicker();

## 5. tabs

- ¿Qué es?
- tabs es un componente que organiza el contenido en pestañas.

- ¿Qué hace?
- Crea una serie de pestañas que el usuario puede navegar. Cada pestaña muestra un contenido diferente sin recargar la página.

- ¿Para qué sirve?
- Es ideal para organizar grandes cantidades de información en un espacio reducido, mejorando la navegación sin hacer scroll.

#### Ejemplo de uso:

        $("#miElementoTabs").tabs();
