# Funciones AJAX en JavaScript

- AJAX (Asynchronous JavaScript and XML) es una técnica que permite que una página web se comunique con el servidor sin necesidad de
  recargarla.
- Esto permite actualizar solo partes específicas de una página y mejora la experiencia del usuario.


## Funciones 

1. **.load()**

- La función .load() permite cargar contenido de una URL y mostrarlo en un elemento HTML específico.

- Sintaxis

        
        $("#elemento").load("url");
- Ejemplo

        $("#contenido").load("pagina.html");
  
- Este código carga el contenido de "pagina.html" y lo muestra dentro del elemento con id contenido.

2. **.get()**

- La función .get() se usa para hacer una solicitud HTTP de tipo GET al servidor.
- Es útil cuando queremos obtener datos (como texto o JSON) de una URL.

- Sintaxis

        $.get("url", function(data){
            // manejar los datos recibidos
        });

- Ejemplo

        $.get("usuarios.json", function(data) {
            console.log(data);
        });

- Este código solicita los datos de "usuarios.json" y los imprime en la consola.

3. **.post()**

- La función .post() envía datos al servidor usando una solicitud HTTP POST.
- Es útil cuando necesitamos enviar datos (como formularios).

- Sintaxis

        $.post("url", { datos }, function(respuesta){
            // manejar la respuesta
        });

- Ejemplo

        $.post("procesar.php", { nombre: "Juan", edad: 30 }, function(respuesta) {
            console.log(respuesta);
        });
  
- Este código envía un objeto con nombre y edad al archivo procesar.php y muestra la respuesta en la consola.

4. **preventDefault()**

- La función preventDefault() evita la acción predeterminada de un evento.
- Es útil, por ejemplo, para evitar que un formulario se envíe al hacer clic en el botón de enviar.

- Ejemplo

        $("form").on("submit", function(event) {
            event.preventDefault();
            alert("Envio del formulario detenido");
        });

  
- En este caso, el formulario no se enviará al hacer clic en "Enviar", y en su lugar se muestra un mensaje de alerta.

5. **$ajax()**

- La función $ajax() permite realizar una solicitud AJAX personalizada y ofrece control total sobre la configuración de la solicitud, como el tipo de método (GET o POST), URL, datos enviados, etc.

- Sintaxis

        $.ajax({
            url: "url",
            type: "GET/POST",
            data: { datos },
            success: function(respuesta) {
                // manejar la respuesta
            }});
  
- Ejemplo

        $.ajax({
            url: "procesar.php",
            type: "POST",
            data: { nombre: "Juan", edad: 30 },
            success: function(respuesta) {
                console.log(respuesta);
            }
        });
  
- Este código envía un objeto con nombre y edad al archivo procesar.php y muestra la respuesta en la consola al recibirla.
