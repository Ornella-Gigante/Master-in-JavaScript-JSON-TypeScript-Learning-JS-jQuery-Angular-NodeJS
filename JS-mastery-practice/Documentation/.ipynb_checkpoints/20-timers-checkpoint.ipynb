# Timers en JavaScript

- En JavaScript, los timers son herramientas que permiten ejecutar código después de un intervalo de tiempo específico.
- Se utilizan comúnmente para realizar acciones repetitivas o retrasadas en una aplicación web.
- JavaScript proporciona dos funciones principales para trabajar con timers: setTimeout() y setInterval().

1. setTimeout()
La función setTimeout() se utiliza para ejecutar una función o un bloque de código después de que haya transcurrido un tiempo específico, medido en milisegundos.

- Sintaxis
        
        setTimeout(función, tiempoEnMilisegundos);
  
- función: La función que deseas ejecutar después de que transcurra el tiempo.
- tiempoEnMilisegundos: El tiempo que deseas esperar antes de ejecutar la función.

- Ejemplo de setTimeout()

        console.log("Inicio");
        
        setTimeout(function() {
            console.log("Esto se ejecuta después de 3 segundos");
        }, 3000);
        
        console.log("Fin

        
- Salida esperada:

        
        Inicio
        Fin
        Esto se ejecuta después de 3 segundos

  
2. setInterval()

- La función setInterval() se utiliza para ejecutar repetidamente una función o un bloque de código en intervalos de tiempo específicos, también medidos en milisegundos.

- Sintaxis

        setInterval(función, tiempoEnMilisegundos);

  
función: La función que deseas ejecutar repetidamente.
tiempoEnMilisegundos: El intervalo de tiempo en milisegundos entre cada ejecución de la función.


- Ejemplo de setInterval()
        
        let contador = 0;
        
        const intervalo = setInterval(function() {
            console.log("Este mensaje aparece cada 2 segundos");
            contador++;
        
            // Detener el intervalo después de 5 repeticiones
            if (contador === 5) {
                clearInterval(intervalo);
                console.log("El intervalo ha sido detenido");
            }
        }, 2000);
  
- Salida esperada:

    
        Este mensaje aparece cada 2 segundos
        (esto se repetirá 5 veces)
  
        El intervalo ha sido detenido

  
3. Cancelación de Timers

Para cancelar un timer creado con setTimeout(), se puede usar clearTimeout().
Para cancelar un timer creado con setInterval(), se puede usar clearInterval().

- Ejemplo de Cancelación

        const timeoutId = setTimeout(() => {
            console.log("Este mensaje no se verá porque fue cancelado");
        }, 5000);
        
        clearTimeout(timeoutId); // Cancela el timeout
  
## Conclusión
Los timers en JavaScript, a través de setTimeout() y setInterval(), son herramientas muy útiles para controlar el tiempo y ejecutar código de manera asíncrona. Permiten crear experiencias interactivas, gestionar eventos en el tiempo y ejecutar tareas repetitivas de manera eficiente.
