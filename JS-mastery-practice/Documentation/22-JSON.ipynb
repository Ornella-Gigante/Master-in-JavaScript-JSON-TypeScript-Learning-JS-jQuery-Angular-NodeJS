# ¿Qué es JSON en JavaScript?

- JSON (JavaScript Object Notation) es un formato de texto que se utiliza para almacenar y transportar datos.
- Aunque tiene "JavaScript" en su nombre, es independiente del lenguaje, pero es especialmente popular en JavaScript porque se parece mucho a los objetos y arreglos de JavaScript.

## ¿Para qué sirve?

- Sirve para intercambiar datos entre un servidor y un navegador (o entre diferentes aplicaciones).
- Es fácil de leer y escribir tanto para humanos como para las máquinas, por eso se usa mucho en APIs y en la comunicación entre sistemas.

## ¿Qué hace?

- Convierte objetos y arreglos de JavaScript a texto para que puedan enviarse o almacenarse.
- Convierte texto en objetos o arreglos de JavaScript para que puedas trabajar con esos datos en tu programa.


## Ejemplo sencillo de JSON:
Aquí tienes un objeto en formato JSON:
        
        {
          "nombre": "Juan",
          "edad": 25,
          "ciudad": "Madrid",
          "amigos": ["Carlos", "Ana", "Lucía"]
        }
        
- Este JSON representa los datos de una persona: su nombre, edad, ciudad, y una lista de amigos.

## Cómo usar JSON en JavaScript

1. Convertir un objeto JavaScript a JSON:
Imagina que tienes un objeto JavaScript y lo quieres convertir a JSON para enviarlo a un servidor.


        const persona = {
          nombre: 'Juan',
          edad: 25,
          ciudad: 'Madrid'
        };
        
        const personaJSON = JSON.stringify(persona);
        console.log(personaJSON); // {"nombre":"Juan","edad":25,"ciudad":"Madrid"}
        El método JSON.stringify() convierte el objeto persona en un texto JSON.

2. Convertir JSON a un objeto JavaScript:
Si recibes un texto en formato JSON y lo quieres usar en tu código como un objeto JavaScript:

        
        const datosJSON = '{"nombre":"Juan","edad":25,"ciudad":"Madrid"}';
        
        const persona = JSON.parse(datosJSON);
        console.log(persona.nombre); // "Juan"
        El método JSON.parse() convierte el texto JSON de vuelta a un objeto JavaScript que puedes usar.

## Resumen:

- JSON es un formato de texto para intercambiar datos.
- JSON.stringify() convierte objetos JavaScript a texto JSON.
- JSON.parse() convierte texto JSON en objetos JavaScript.
- Es muy útil para enviar y recibir datos entre aplicaciones o entre cliente y servidor.
