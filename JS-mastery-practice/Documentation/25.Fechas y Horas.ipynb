
# Fecha y Hora en JavaScript

- En JavaScript, el objeto `Date` permite trabajar con fechas y horas.
- A continuación, se muestran los métodos más comunes para obtener partes específicas de una fecha.

## Crear un Objeto de Fecha


        let fecha = new Date();
        
- Crea un objeto de fecha que representa el momento actual (fecha y hora).

### Métodos Principales de Date

1. getMonth()
   
- Descripción: Obtiene el mes de la fecha.
- Valores: 0 para enero, 1 para febrero, ..., 11 para diciembre.
- Ejemplo:

        fecha.getMonth(); // Si es octubre, devuelve 9

   
2. getDate()

- Descripción: Obtiene el día del mes.
- Valores: 1 a 31.
- Ejemplo:

        fecha.getDate(); // Si es 31 de octubre, devuelve 31

  
4. getDay()

- Descripción: Obtiene el día de la semana.
- Valores: 0 para domingo, 1 para lunes, ..., 6 para sábado.
- Ejemplo:

        fecha.getDay(); // Si es jueves, devuelve 4

  
5. getHours()

- Descripción: Obtiene la hora en la zona horaria local.
- Valores: 0 a 23 (formato de 24 horas).
- Ejemplo:

        fecha.getHours(); // Si son las 14:00, devuelve 14

  
6. getUTCHours()

- Descripción: Obtiene la hora en UTC (Tiempo Universal Coordinado).
- Valores: 0 a 23.
- Ejemplo:

        fecha.getUTCHours(); // Si son las 14:00 (UTC+2), devuelve 12

  
7. getMinutes()
   
- Descripción: Obtiene los minutos en la zona horaria local.
- Valores: 0 a 59.
- Ejemplo:
        
        fecha.getMinutes(); // Si son las 14:30, devuelve 30

  
# Ejemplo Completo
        
        let fecha = new Date();
        console.log("Mes:", fecha.getMonth());          // Ejemplo: 9 para octubre
        console.log("Día del mes:", fecha.getDate());   // Ejemplo: 31
        console.log("Día de la semana:", fecha.getDay()); // Ejemplo: 4 (jueves)
        console.log("Hora local:", fecha.getHours());   // Ejemplo: 14 (2 PM)
        console.log("Hora UTC:", fecha.getUTCHours());  // Ejemplo: 12 (UTC)
        console.log("Minutos:", fecha.getMinutes());    // Ejemplo: 30
